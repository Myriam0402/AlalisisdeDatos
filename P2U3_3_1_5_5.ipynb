**Observe la diagonal de izquierda a derecha en la tabla de correlación generada arriba. ¿Por qué la diagonal está plagada de 1? ¿Se trata de una coincidencia? Explique.** La presencia de valores de 1 en la diagonal principal de una matriz de correlación no es una coincidencia, sino una característica fundamental de esta representación matricial.

La diagonal principal de una matriz de correlación muestra la correlación de cada variable consigo misma. Como es lógico, cualquier variable tiene una correlación perfecta consigo misma, lo que se traduce en un valor de 1 en la diagonal principal.

Esto es consistente con la definición matemática de correlación. Cuando calculamos la correlación de una variable consigo misma, el resultado es siempre 1, ya que la variable está perfectamente relacionada consigo misma.

Por lo tanto, la presencia de 1 en la diagonal principal de una matriz de correlación no es una coincidencia, sino una consecuencia natural de cómo se calcula y se representa la correlación entre variables.


**Si continuamos observando la tabla de correlación que figura arriba, notaremos que los valores se duplican; los valores debajo de la diagonal 1 tienen valores homólogos duplicados sobre la diagonal 1. ¿Se trata de una coincidencia? Explique.**
No, no es una coincidencia. Lo que pasa es típico en las matrices de correlación. Estas matrices suelen ser simétricas con respecto a su diagonal principal. La razón de los valores duplicados arriba y debajo de la diagonal principal es porque la correlación entre la variable X e Y es la misma que la correlación entre Y e X.

**Muchos pares de variables presentan una correlación cercana a cero. ¿Qué significa?**

Una correlación cercana a cero entre pares de variables indica que no hay una relación lineal fuerte entre ellas. En otras palabras, los cambios en una variable no están relacionados o no predicen los cambios en la otra variable de manera consistente. Sin embargo, es importante destacar que una correlación cercana a cero no significa necesariamente que no haya ninguna relación entre las variables, ya que podrían existir relaciones no lineales o asociaciones que no se capturan mediante la correlación lineal.
