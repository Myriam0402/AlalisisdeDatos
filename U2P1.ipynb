```python
#Reemplazar las celdas cuyo dato sea "?" por "nan"
#que significa "not a number"
#atrubuto "inplace=True" hace el cambio permanente en
#el archivo CSV original
#ya que por default solo lo hace en memoria pero no
#modifica el archivo fuente
autos.replace("?",np.nan, inplace= True)
autos
```

```python
#¿Qué hacer con los datos faltantes?
#Sacaremos el promedio en las columnas cuyos datos sean numéricos
promedio_normalized_losses = autos['normalized-losses'].astype("float").mean(axis=0)
print(promedio_normalized_losses)
```

```python
#Reemplazar los valores nulos por el promedio que nos salió
#que fue 122
autos['normalized-losses'].replace(np.nan, promedio_normalized_losses, inplace = True)
```

```python
#Ahora hacemos eso con todas las columnas que tengan valores nulos
#La siguiente es num-of-doors, y aunque sí son números, es un valor categórico
#o descriptivo del auto (cuántas puertas tiene) para ello, no sacaremos el
#promedio sino la moda. Para ello calcularemos cuál valor se repite más
autos["num-of-doors"].value_counts()
```

```python
#El valor que más se repitió fue 4, entonces hacemos la sustitución
autos['num-of-doors'].replace(np.nan, "four", inplace = True)
```

```python
#La siguiente columna es "bore"
promedio_bore = autos['bore'].astype("float").mean(axis=0)
print("Promedio de columna bore: ",promedio_bore)
#Sustituyendo los valores por el promedio
autos['bore'].replace(np.nan, promedio_bore, inplace = True)
```

```python
#Columna horsepower
promedio_horsepower = autos['horsepower'].astype("float").mean(axis=0)
print("Promedio de columna horsepower: ",promedio_horsepower)
#Sustituyendo los valores por el promedio
autos['horsepower'].replace(np.nan, promedio_horsepower, inplace = True)
```

```python
#Columna peak-rpm
promedio_peakrpm = autos['peak-rpm'].astype("float").mean(axis=0)
print("Promedio de columna peak-rpm: ",promedio_peakrpm)
#Sustituyendo los valores por el promedio
autos['peak-rpm'].replace(np.nan, promedio_peakrpm, inplace = True)
```

```python
#CORREGIR EL FORMATO DE LOS DATOS
#En pandas el siguiente código es para revisar el tipo de datos
autos.dtypes
#El siguiente código es para cambiar el tipo de datos
#dataframe.astype()
```

```python
#Para estandarizar los datos creamos una nueva columna de conversión para
#unidades más familiares (el consumo de combustible a l/100km)
#Fórmula de conversión:
#L/100Km = 235/ m.p.g
autos['city-L/100Km'] = 235 / autos['city-mpg']
#Revisar la tabla a ver si se hizo el cambio
autos.head()
```

```python
#Editar columna para convertir unidades de millas por galón (mpg)
#a unidades más familiares (litros por 100 kilómetros)
#Fórmula de conversión:
#L/100Km = 235/ m.p.g
#Ahora no creamos nueva columna sino que editamos la existente
autos['highway-mpg'] = 235 / autos['highway-mpg']
#Como editamos la columna, la renombraremos de "highway-mpg" a
#"highway-L/100km"
autos.rename(columns={"highway-mpg":"highway-L/100km"}, inplace=True)
#Checar datos transformados
autos.head()
```

```python
#Primero eliminamos la columna que habíamos creado
autos.drop(['city-L/100Km'], axis=1, inplace=True)
#axis 1 se refiere a columnas
#axis 0 se refiere a filas
#Ahora editamos la columna 'city-mpg' para darle los valores estandarizados
autos["city-mpg"] = 235 / autos["city-mpg"]
autos.rename(columns={"city-mpg":"city-L/100km"}, inplace=True)
#Confirmamos el cambio
autos.head()
```

```python
#NORMALIZACIÓN DE LOS DATOS
#Visualizamos las columnas a normalizar
autos[["length","width","height"]].head()
```

```python
#Intercambiamos los valores originales por: valor original / valor máximo
autos["length"] = autos["length"] / autos["length"].max()
autos["width"] = autos["width"] / autos["width"].max()
autos["height"] = autos["height"] / autos["height"].max()
#Verificamos los cambios
autos[["length","width","height"]].head()
```
