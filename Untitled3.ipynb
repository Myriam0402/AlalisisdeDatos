```python
#Contar por grupo cuántos juegos son por
#plataforma y genero únicamente Acción y Aventura
Ejercicio1=juegos[((juegos["Genre"]=="Action")|(juegos["Genre"]=="Adventure"))]
Ejercicio1.groupby(["Plataforma","Genre"]).count()
```

```python
#Contar por grupos cuántos juegos son por género y año únicamente
#de los años 2010 y superiores
Ejercicio2=juegos[((juegos["Year"]>=2010))].groupby(["Genre","Year"]).count()
Ejercicio2
```

```python
#Agrupar y sumar las ventas globales por género
#sum en ventas globales
Ejercicio3= juegos[["Global_Sales","Genre"]].groupby("Genre")["Global_Sales"].sum()
Ejercicio3
```
